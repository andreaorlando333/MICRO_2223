**Esercizi con MATLIB**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
stelle=30
X= np.random.uniform(0.,1.,stelle)
Y= np.random.uniform(0.,1.,stelle)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.set_facecolor("black")
ax.plot(X,Y,marker = "*",markersize=16,linestyle=":",color="yellow")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
s1=np.sin(x)
s2=np.cos(x)
```

```python
fig, ax  =plt.subplots(figsize=(10,5))
ax.plot(x,s1,linestyle=":",linewidth = 2,color="red",label="seno")
ax.plot(x,s2,linestyle="-.",linewidth = 2,color="blue",label="coseno")
ax.set_xlabel("x",fontsize=12)
ax.set_ylabel("y",fontsize=12)
ax.set_title("Grafico generato:",fontsize=12)
ax.tick_params(axis = "both",labelsize=20)
plt.legend()
plt.savefig("./grafici.pdf")
plt.show()
```

```python
import numpy as np
n = np.random.randint(600,800)

blu_scuro = (0,0,24/255)
colori_stelle = (np.random.uniform(220,240)/255,np.random.uniform(240,255)/255,np.random.uniform(155,255)/255)

x = np.random.uniform(0.,1.,n)
y = np.random.uniform(0.,1.,n)
colori = np.random.uniform(0.,1.,(n,3))
dimensione = np.random.uniform(1.,15.,n)
trasparenza = np.random.uniform(0.,1.,n)

fig,ax = plt.subplots(figsize=(10,10))
ax.scatter(x,y,c = colori_stelle, s = dimensione, alpha = trasparenza)
ax.set_facecolor(blu_scuro)
plt.xticks([])
plt.yticks([])
ax.grid(False)
plt.show()
```
