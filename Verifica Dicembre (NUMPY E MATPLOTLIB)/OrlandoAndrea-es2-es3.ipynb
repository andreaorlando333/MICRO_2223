```python
import numpy as np
import matplotlib.pyplot as plt
```

**Esercizio 2**, prezzi giornalieri di un'azione:

```python
n_days = 200
np.random.seed(121)
prezzo_giornaliero = 100 + np.cumsum(np.random.normal(0.1,1.0,n_days))
# 1
valoreMassimo = max(prezzo_giornaliero)
valoreMinimo = min(prezzo_giornaliero)
print(valoreMassimo, valoreMinimo)

# 2
counter = 0
for i in prezzo_giornaliero:
    if i > 105.0:
        counter = counter + 1
print(f"Giorni con prezzo maggiore di 105: {counter}")

# 3
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(prezzo_giornaliero)
plt.title("Andamento prezzi:")
plt.show()

# 4
variazioneMax = 0
contaGiorni = 0
giorno = 0
for i in prezzo_giornaliero:
    contaGiorni = contaGiorni + 1
    giornoSuccessivo = i + 1 
    variazione = i - giornoSuccessivo
    
    if variazione < variazioneMax:
        variazioneMax = variazione
        giorno = contaGiorni
        
print(giorno)
```

**Esercizio 3**, funzioni:

```python
x = np.linspace(-3*np.pi,3*np.pi, 600)
f1 = np.exp(-x**2)
f2 = np.exp(-x**2)*np.cos(20*x)
f3 = np.exp(-x**2)*np.cos(20*x)
f4 = (np.sin(20*x))/(20*x)

plt.plot(x,f1,label='f1')
plt.plot(x,f2,label='f2')
plt.plot(x,f3,label='f3')
plt.plot(x,f4,label='f4')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
