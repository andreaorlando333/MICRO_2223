```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("emotion.csv")
righe = file.readlines()
file.close()
```

```python
dati = {"emozioni":[], "pixel":[]}
```

```python
for riga in righe[1:]:
    colonne = riga.split(",")
    dati["emozioni"].append(colonne[0])
    stringa_img = colonne[1][:-1]
    lista_immagine = [int(valore)
                     for valore in stringa_img.split(" ")]
    img = np.array(lista_immagine).reshape(48,48)
    dati["pixel"].append(img)
```

```python
fig, axs = plt.subplots(nrows = 3, ncols = 3, figsize = (6,6))
cnt = 0
for i in range(0,3):
    for j in range(0,3):
        axs[i,j].imshow(dati["pixel"][cnt], cmap = "gray")
        axs[i,j].set_axis_off()
        cnt = cnt + 1
        
plt.show()
```
