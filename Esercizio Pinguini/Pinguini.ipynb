```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
with open("penguins_size.csv", newline = "", encoding = "utf8") as filecsv:
    lettore = csv.reader(filecsv, delimiter = ";")
    header = next(lettore)
    # Stampo le intestazioni (keys)
    print(header)
```

```python
f = open("penguins_size.csv", "r")
righe = f.readlines()
f.close()

dati = {"species":[],"island":[],"culmen_length_mm":[],
        "culmen_depth_mm":[],"flipper_length_mm":[],
        "body_mass_g":[],"sex":[]}

for riga in righe[1:]:
    lista_campi = riga[:-1].split(",")
    if "NA" not in lista_campi:
        dati["species"].append(lista_campi[0])
        dati["island"].append(lista_campi[1])
        dati["culmen_length_mm"].append(float(lista_campi[2]))
        dati["culmen_depth_mm"].append(float(lista_campi[3]))
        dati["flipper_length_mm"].append(float(lista_campi[4]))
        dati["body_mass_g"].append(float(lista_campi[5]))
        dati["sex"].append(lista_campi[6])
```

```python
# Trasformo le liste del dizionario in array NUMPY
for nome_colonna, lista in dati.items():
    dati[nome_colonna] = np.array(lista)
```

Creo un istogramma per **confrontare** le specie di pinguini rilevate:

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(dati["species"], color = "orange")
plt.show()
```

Cerco una **correlazione** tra la lunghezza della pinna e il peso del pinguino:

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(dati["body_mass_g"], dati["flipper_length_mm"])
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].hist(dati["body_mass_g"], color = "red")
axs[1].scatter(dati["body_mass_g"], dati["flipper_length_mm"], color = "Orange")
axs[0].set_title("Grafico generato:",fontsize=12)
plt.show()
```
