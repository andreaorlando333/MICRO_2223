```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml("mnist_784")
```

```python
data_array = mnist['data'].values
# Stampo tipologia e dimensioni
print(type(data_array))
print(data_array.shape)
```

```python
esempio_cifra = data_array[1234].reshape((28,28))
print(esempio_cifra.shape)
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.imshow(esempio_cifra, cmap = "Greys")
plt.show()
```

```python
nrows = 5
ncols = 5
fig, axs = plt.subplots(nrows = nrows, ncols = ncols, figsize=(5,5))
for i in range(nrows):
    for j in range(ncols):
        esempio_cifra = data_array[np.random.randint(0,70000)].reshape((28,28))
        axs[i,j].imshow(esempio_cifra, cmap = "Greys")
        axs[i,j].set_axis_off()
plt.show()
```
